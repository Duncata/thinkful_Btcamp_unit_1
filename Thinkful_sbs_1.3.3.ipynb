A box plot (or box-and-whisker plot) shows the distribution of quantitative
data in a way that facilitates comparisons between variables or across
levels of a categorical variable. The box shows the quartiles of the
dataset while the whiskers extend to show the rest of the distribution,
except for points that are determined to be "outliers" using a method
that is a function of the inter-quartile range.

Show in the boxplot is that there is no outlier from the data column Residential Revenue. The  Median in the Residential Revenue is in the 10 Million. The quartiles of the residential revenue columns are as follows; 0%-25% of the data value w/ in 7 Million and 8 million, 25%-50% of the data value w/ in 8 Million and 10 million, 50%-75% of the data value w/ in 10 Million and 13 million, 75%-100% of the data value w/ in 13 Million and 15 million. Some of the disadvantage of boxplot as shown below is tha that it can consume space.

Box and whisker plots handle large data effortlessly, but they do not retain the exact values and the details of the results of the distribution. These graphs allow a clear summary of large amounts of data.



```python
%matplotlib inline
sns.boxplot(data=df_rev, y=df_rev['Residential'])
plt.title('RESIDENTIAL REVENUE DATA BOXPLOT')
```

A bar plot represents an estimate of central tendency for a numeric
variable with the height of each rectangle and provides some indication of
the uncertainty around that estimate using error bars. Bar plots include 0
in the quantitative axis range, and they are a good choice when 0 is a
meaningful value for the quantitative variable, and you want to make
comparisons against it.

For datasets where 0 is not a meaningful value, a point plot will allow you
to focus on differences between levels of one or more categorical
variables.

In the barplot shown below, the height of the bar shows the count value which is in the range of 10 Million. The vertical line on top of it represents the standard deviation (or standard error)

A pointplot has all the strengths of a barplot without many of its weaknesses.  Pointplots use much less space to convey the same information, and the difference in size between the groups is rightly compared to the size of the error bars.

The point plot shows only the mean (or other estimator) value, but in many cases it may be more informative to show the distribution of values at each level of the categorical variables.

A violin plot plays a similar role as a box and whisker plot. It shows the
distribution of quantitative data across several levels of one (or more)
categorical variables such that those distributions can be compared. Unlike
a box plot, in which all of the plot components correspond to actual
datapoints, the violin plot features a kernel density estimation of the
underlying distribution.

This can be an effective and attractive way to show multiple distributions
of data at once, but keep in mind that the estimation procedure is
influenced by the sample size, and violins for relatively small samples
might look misleadingly smooth.

A Violin Plot is used to visualise the distribution of the data and its probability density. This chart is a combination of a Box Plot and a Density Plot Widths(=frequency) that is rotated and placed on each side, to show the distribution shape of 
the data.

### 3. Choose one continuous variable and one categorical variable, and plot them six different ways.

```python
df_LCDUS= pd.read_csv('Downloads/NCHS_-_Leading_Causes_of_Death__United_States.csv')
df_LCDUS.drop('113 Cause Name', axis=1, inplace=True)
df_LCDUS.columns = df_LCDUS.columns.str.replace(' ', '_').str.replace('-', '_').str.lower()
df_LCDUS['population'] = (df_LCDUS['deaths']*100000) / df_LCDUS['age_adjusted_death_rate']
df_diseases = df_LCDUS[(df_LCDUS.cause_name != 'All causes') & 
                        (df_LCDUS.cause_name !='Unintentional injuries') & 
                        (df_LCDUS.cause_name !='Suicide')]
df_LCDUS_clean = df_diseases[df_diseases.state != 'United States']
df_LCDUS_clean.head()
```

The lineplot shows all the continous value plot of categorical diseases in x axis color coded by year. 
Heart disease has the high peak in deaths and kidney shows the lowest value in deaths

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.lineplot(x='cause_name', y='deaths', hue='year', data = df_LCDUS_clean)
plt.xticks(rotation=70)
```

The height of a bar indicates the number of deaths, it shows clearly in this bar plot that heart disease shows the 
highest height and kidney disease shows the lowest height.The vertical line on top of it represents the standard deviation (or standard error).

```python
plt.rcParams['figure.figsize'] = 10, 5
ax = sns.barplot(x='cause_name', y='deaths', data = df_LCDUS_clean)
plt.xticks(rotation=70)
```

In this plot it it value is represented by points and colors by year.
It is the same w/ bar plot but bar plot visualize more clearly. 

```python
plt.rcParams['figure.figsize'] = 10, 7
ax = sns.scatterplot(x='cause_name', y='deaths', hue='year', data = df_LCDUS_clean)
plt.xticks(rotation=90)
```

The point plot shows only the mean (or other estimator) value of the all kind of diseases, but in this case bar plot shows more information visually with the distribution of values at each level of the categorical variables.

```python
# Setting the overall aesthetic.
plt.rcParams['figure.figsize'] = 2, 4
sns.set(style="whitegrid")

g = sns.factorplot(x='cause_name', y="deaths", data= df_LCDUS_clean,
                   size=12, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("DEATHSs")
g.set_xlabels("")
plt.title('Pointplot: DEATH PER Disease')
plt.xticks(rotation=70)
plt.show()
```

A marginal plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot. ... The seaborn library provides a joint plot function that is really handy to make this type of graphic.

Seaborn - Kernel Density Estimates. Advertisements. Kernel Density Estimation (KDE) is a way to estimate the probability density function of a continuous random variable. It is used for non-parametric analysis.

```python
sns.jointplot('Attack', 'Defense' , kind='kde', size =7,space=0, color='g', data= df_poke)
```

```python
sns.jointplot('Attack', 'Defense' , kind='reg', size =7,space=0, color='g', data= df_poke)
```

```python
sns.jointplot('Attack', 'Defense' , kind='hex', size =7,space=0, color='g', data= df_poke)
```

```python
from scipy.stats import spearmanr
sns.jointplot('Attack', 'Defense' , color='g', data= df_poke, stat_func=spearmanr)
```
